```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edge("Victim", "Mule_A")
G.add_edge("Mule_A", "Mule_B")
G.add_edge("Mule_A", "Mule_C")
G.add_edge("Mule_B", "Cash_Out")

plt.figure(figsize=(10,8))

colors = ["skyblue","red","red","red","green"]

pos = {
    "Victim": (2,2),
    "Mule_A": (1,1),
    "Mule_B": (0,0),
    "Mule_C": (2,0),
    "Cash_Out": (1,-1)
}

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=colors,
    node_size=4000,
    font_size=10,
    arrows=True
)

plt.title("Graph-Based Mule Account Detection")
plt.axis("off")
plt.savefig("mule_network_final.png", bbox_inches="tight")
plt.show()
```
